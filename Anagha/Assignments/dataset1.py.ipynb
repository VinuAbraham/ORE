# 3) Missing value treatment

```python
df.isnull()
```

```python
sns.heatmap(df_feature.isnull(),yticklabels=False)
```

```python
 df['LSTAT']=df['LSTAT'].fillna((df['LSTAT'].mode()[0]))
```

```python
 df['PTRATIO']=df['PTRATIO'].fillna((df['PTRATIO'].mode()[0]))
```

```python
df.isnull()
```

```python
min_thresold, max_thresold = df_feature.ZN.quantile([0.001, 0.999])
min_thresold, max_thresold
df2 = df[(df_feature.ZN<max_thresold) & (df_feature.ZN>min_thresold)]
plt.hist(df2['ZN'])
plt.show()
```

```python
min_thresold, max_thresold = df_feature.INDUS.quantile([0.001, 0.999])
min_thresold, max_thresold
df3 = df[(df_feature.INDUS<max_thresold) & (df_feature.INDUS>min_thresold)]
plt.hist(df3['INDUS'])
plt.show()
```

```python
min_thresold, max_thresold = df.PTRATIO.quantile([0.001, 0.999])
min_thresold, max_thresold
df4 = df[(df.PTRATIO<max_thresold) & (df.PTRATIO>min_thresold)]
plt.hist(df4['NOX'])
plt.show()
```

```python
min_thresold, max_thresold = df.LSTAT.quantile([0.001, 0.999])
min_thresold, max_thresold
df5 = df[(df.LSTAT<max_thresold) & (df.LSTAT>min_thresold)]
plt.hist(df5['RM'])
plt.show()
```
